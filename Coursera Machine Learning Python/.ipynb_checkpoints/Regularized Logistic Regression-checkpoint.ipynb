```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt("ex2data1.txt",delimiter=",")
X=data[:,[0,1]]
y=data[:,2]
y=y.reshape((100,1))
```

```python
#plotting the data

plt.scatter(X[:,0],X[:,1],c=y,marker="+")
```

```python
def sigmoid(z):
    final=1./(1+np.exp(-z))
    return final

print sigmoid(np.array([-10,0,10]))
```

```python
theta=np.zeros((2,1),dtype="float")
def costFunction(theta,X,y):
    
    m=X.shape[0]  #this is the number of training examples
    theta=theta.reshape((2,1))   #i have to put this because fmin returns a numpy array,converting array to a matrix 
    

    
    hx=np.dot(X,theta)
    hx=sigmoid(hx)
    
    cost=np.sum((-y*np.log(hx) - (1-y)*np.log(1-hx)),axis=0)
    
    
    cost=(1./m) * cost
    
    #calculating the gradient of the log likelohood cost function
    
    grad=np.sum(((sigmoid(hx)-y)*X),axis=0)
    grad=(1./m) * grad
    grad=grad.reshape((2,1))
    return cost

costFunction(theta,X,y)
```

Now i can write the gradient descent update rule for this optimisation problem but in a lot of optimisation problem the optimisation to be carried is quite complicated. Scipy and numpy have optimsation routine built in them using these routine i can find the minima of the function.

```python
from scipy.optimize import fmin
fmin(costFunction,np.zeros((2,1),dtype="float"),args=(X,y),maxiter=100)
```
