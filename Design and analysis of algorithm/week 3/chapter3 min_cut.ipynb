```python
#so now i have extracted the data , now i need to plan the routines needed to solve this problem

#first solve this problem on paper with pseudo code and then get to coding.

''' the functions needed to solve this problem.
1. a function for picking up the correct number of edges.     i will have to maintain a set of correct edges
2. a functino for merging 
3. a function for removing the self loop


the graph G is stored as a dictionary
'''
```

```python
import random
def choose_node(set_of_nodes):
    node1,node2=random.sample(set_of_nodes,2)
    return node1,node2

def merge_nodes(graph,node1,node2,set_of_nodes):
    neighbours_node1=set(graph[node1])
    
    #adding the elements to the node1
    for elem in graph[node2]:
        neighbours_node1.add(elem)
    graph[node1]=list(neighbours_node1)
    
    #for all the nodes that point to node 2 i am making them point towards node1
    
    for key in graph.keys():
        if key==node1 or key==node2:
            continue
        elif node2 not in graph[key]:
            continue
        
        temp_lst=graph[key]
        
        final_lst=[elem for elem in temp_lst if elem!=node2]
        final_lst.append(node1)
        
        graph[key]=final_lst
        
    #removing the node2 from the graph
    
    del graph[node2]
    
    #removing node2 from the set_of_nodes
    
    set_of_nodes.remove(node2)
    
    return graph,set_of_nodes


def remove_self_loop(graph,node):
    temp_lst=[elem for elem in graph[node] if elem!=node]
    graph[node]=temp_lst
    return graph
        
        
        
    
    
```
